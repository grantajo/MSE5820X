```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

## 10.11 Lognormal Distribution

THe figure below shows a histogram of a dataset that was sampled following a lognormal distribution together with the "true" PDF:

```python
mu = 1
sigma = 0.85

# create a continuous lognorm RV
rv = stats.lognorm(sigma, mu)

# generate lognorm distribtued random values
r = stats.lognorm.rvs(sigma, loc=mu, size=800)
```

```python
fig, ax = plt.subplots()
x = np.linspace(stats.lognorm.ppf(0.01, sigma), stats.lognorm.ppf(0.99, sigma), 100)
ax.hist(r, bins=np.linspace(0, 7, 60), density=True);
ax.plot(x, rv.pdf(x))
ax.set(xlabel='X', ylabel='PDF');
```
