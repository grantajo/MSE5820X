```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mass = 4 # amu
m = mass / 6.022e+26 # convert from amu to kg
kb = 1.38e-23 # J/K
temp = 298.15 # K
pi = np.pi

def f(c):
    first = 4 * pi * c**2
    second = (m / 2 / pi / kb / temp)**1.5
    third = np.exp(-m * c**2 / 2 / kb / temp)
    return first * second * third
```

```python
x = np.linspace(0,2500,2500)
y = f(x)
```

```python
plt.plot(x,y)
plt.xlabel('Speed [m/s]')
plt.ylabel('Probability')
plt.show()
```

```python
index = np.argmax(y)
maximum = x[index]
print('The most probable speed is',maximum,'m/s')
```

```python
def ft(c,temp):
    first = 4 * pi * c**2
    second = (m / 2 / pi / kb / temp)**1.5
    third = np.exp(-m * c**2 / 2 / kb / temp)
    return first * second * third

temp0 = 100
temp1 = 300
temp2 = 400
temp3 = 500

y0 = ft(x,temp0)
y1 = ft(x,temp1)
y2 = ft(x,temp2)
y3 = ft(x,temp3)
```

```python
plt.plot(x,y, label="100 K")
plt.plot(x,y1, label="300 K")
plt.plot(x,y2, label="400 K")
plt.plot(x,y3, label="500 K")
plt.legend()
plt.xlabel('Speed [m/s]')
plt.ylabel('Probability')
plt.show()
```

```python
def fm(c,mass):
    m = mass / 6.022e+26
    first = 4 * pi * c**2
    second = (m / 2 / pi / kb / temp)**1.5
    third = np.exp(-m * c**2 / 2 / kb / temp)
    return first * second * third

yHe = fm(x,4)
yNe = fm(x,20)
yAr = fm(x,40)
yXe = fm(x,131)
```

```python
plt.plot(x,yHe, label="He")
plt.plot(x,yNe, label="Ne")
plt.plot(x,yAr, label="Ar")
plt.plot(x,yXe, label="Xe")
plt.legend()
plt.xlabel('Speed [m/s]')
plt.ylabel('Probability')
plt.show()
```
